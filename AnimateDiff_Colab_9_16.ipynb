```python
# Import required packages
from PIL import Image
import torch
from diffusers import AnimateDiffPipeline
import moviepy.editor as mpy
```

```python
# Load pipeline and set prompt
pipe = AnimateDiffPipeline.from_pretrained("guoyww/animatediff")
pipe.to("cuda")
prompt = "a colorful fantasy jungle with magical animals, ultra hd, smooth camera motion"
```

```python
# Generate animation (16:9 is default; we will crop to 9:16)
video_frames = pipe(prompt=prompt, num_frames=24, height=1024, width=576).frames
video_clip = mpy.ImageSequenceClip(video_frames, fps=12)
video_clip.write_videofile("animated_video.mp4")
```
